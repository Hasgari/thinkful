```python
def to_numeric(x):
    x = x.strip()
    x = x[1:]
    if x[-1] == "K":
        return float(x[:-1])
    elif str(x)[-1] == "M":
        return float(x[:-1])*1000
    else:
        return float(x)
```

```python
df["numeric_Wage"]=df["Wage"].apply(lambda x: to_numeric(x))
```

```python
to_numeric(df["Wage"][0])
```

```python
g = sns.lmplot(y = "numeric_Wage", x = "Age", data = df, fit_reg = True, scatter_kws = {"alpha": 0.4})
g.set_ylabels("Wage")
g.set_xlabels("Age")
plt.title("Scatterplot: Wage by Age")
```

```python
g = sns.lmplot(y = "numeric_Wage", x = "Age",hue = "Preferred Foot" , data = df, fit_reg = True, scatter_kws = {"alpha": 0.4})
g.set_ylabels("Wage (1000 Euros)")
g.set_xlabels("Age")
plt.title("Scatterplot: Wage by Age")
```

```python
g = sns.lmplot(y = "numeric_Wage", x = "Age",col = "Preferred Foot" , data = df, fit_reg = True, scatter_kws = {"alpha": 0.4})
g.set_ylabels("Wage (1000 Euros)")
g.set_xlabels("Age")
plt.suptitle("Scatterplot: Wage by Age")
plt.subplots_adjust(top = 0.8)
```

```python
df["numeric_value"] = df["Value"].apply(lambda x : to_numeric(x))
```

```python
Barcelona_value = df.loc[df["Club"] == "FC Barcelona", "numeric_value"]
```

```python
Real_Madrid_value = df.loc[df["Club"] == "Real Madrid", "numeric_value"]
```

```python
plt.hist(Barcelona_value, color = "red", alpha = 0.5,hatch="///", label = "Barcelona Players Value")
plt.hist(Real_Madrid_value, color = "yellow", alpha = 0.5, label = "Real Madrid Players Value")
plt.xlabel("Values in 1000 Euros")
plt.legend(loc = "upper right")
plt.title("Player Value Comparison between Barcelona and Real Madrid")
plt.show()
```

```python
df_barca_real = df[(df["Club"] == "FC Barcelona")|(df["Club"] == "Real Madrid")]
```

```python
df_barca_real["numeric_value"]
```

```python
sns.set(style = "ticks")
sns.set(color_codes = True)
g = sns.FacetGrid(df_barca_real,col = "Club")
g.map = (plt.hist, "numeric_value", color = "red")
plt.subplots_adjust(top = 0.8)
g.set_axis_labels("Values in 1000 Euros")
plt.suptitle("Player Value Comparisons between Two Clubs")
sns.despine(trim = True)
plt.show()
```

```python
sns.set(style = "whitegrid")
g = sns.catplot(x = "Club", y = "numeric_value", data = df_barca_real, height = 6, kind = "point", palette = "pastel", ci = 95)
g.despine(left = True)
g.set_ylabels("Player Value")
g.set_xlabels("")
plt.title("Player Value Comparisons between Two Clubs")
plt.show()
```

```python
sns.set(style = "whitegrid")
g = sns.catplot(x = "Club", y = "numeric_value", hue = "Preferred Foot", data = df_barca_real, height = 6, kind = "point", palette = "pastel", ci = 95, dodge = True, join = False)
g.despine(left = True)
g.set_ylabels("Player Value")
g.set_xlabels("")
plt.title("Player Value Comparisons between Two Clubs")
plt.show()
```

```python
sns.set(style = "whitegrid")
g = sns.catplot(x = "Club", y = "numeric_value", hue = "Preferred Foot", data = df_barca_real, height = 6, kind = "bar", palette = "pastel", ci = 95)
g.despine(left = True)
g.set_ylabels("Player Value")
g.set_xlabels("")
plt.title("Player Value Comparisons between Two Clubs")
plt.show()
```

```python
sns.set(style="whitegrid")
graph = sns.boxplot(x = "Club", y = "numeric_value", hue = "Preferred Foot", data = df_barca_real, palette = "pastel")
plt.title("Player Value by Club and Preferred Foot")
sns.despine(offset = 10, trim = True)
graph.set(xlabel = "", ylabel = "Player Value")
```
